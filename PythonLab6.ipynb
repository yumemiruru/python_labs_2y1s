```python
import math
def task_10(N):
    if not isinstance(N, int) or N < 0:
        return "Помилка. Надане значення повинно бути додатнім числом."

    found = False

    for a in range(1, int(math.sqrt(N)) + 1):
        b_squared = N - a**2

        if b_squared >= 0:
            b = int(math.sqrt(b_squared))
            if b**2 == b_squared:
                return f'{a}^2 + {b}^2 = {N}'
                found = True

    if not found:
        return f'Число {N} не можна представити квадратами цілих чисел'
```

```python
print("Тест 1:", task_10(45))
```

```python
print("Тест 2:", task_10(14))
```
